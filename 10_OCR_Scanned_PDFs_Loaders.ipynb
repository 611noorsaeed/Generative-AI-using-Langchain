# Images, OCR (Scanned PDF Loaders)

# 1. UnstructuredImageLoader

Definition: Loads text from image files (.png, .jpg, etc.) using the Unstructured library with OCR support.
When to use: Use when you need a general OCR solution for images, works for most formats.

```python
from langchain_community.document_loaders import UnstructuredImageLoader
"""
Params:
mode = 'single' 
languages = ['eng']
"""
loader = UnstructuredImageLoader("image.png", languages = ["eng"])
docs = loader.load()

print(docs[0].page_content)  
```

# 2 UnstructuredPDFLoader
Definition: Extracts text from scanned PDFs using the unstructured library (adds OCR layer to the PDF before reading).
When to use: Use for scanned or image-based PDFs where normal PDF loaders fail.

```python
from langchain_community.document_loaders import UnstructuredPDFLoader

loader = UnstructuredPDFLoader("i-9.pdf")
docs = loader.load()
print(docs[0].page_content)
```
